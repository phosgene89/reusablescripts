# Detrending and Deseasoning

Check for trends, typically those described by polynomials. Ideally the fitted line should have an interpretation to avoid overfitting, but this has to be weighted against the descriptive power of the fitted curve.

Then check for seasonal variation.

If either trends or seasonality is detected, remove them before fitting an ARIMA model.

First we will try fitting a line.

````python
from statsmodels.sandbox.regression.predstd import wls_prediction_std

X = np.arange(len(dta))
X_added_constant = sm.add_constant(X) # statsmodels requires us to add constants for regression
y = dta.SUNACTIVITY.values
y = y - np.mean(y)

model = sm.OLS(y, X_added_constant)
results = model.fit()

plt.plot(y)
plt.plot(results.predict(X_added_constant))
````

````python
print(results.summary())
````

````python
print('Parameters: ', results.params)
print('R2: ', results.rsquared)
````

It doesn't look like the line really describes any trends; it looks like it has a non-zero slope due to large outliers later in the series. We can conclude that there is no trend.

We now check for seasonality. To check for seasonality, we do a Fourier transform and check to see if any particular frequencies are significant. For more info on implementing a Fast Fourier Transform (FFT) in Python, click <a href ="https://www.oreilly.com/library/view/elegant-scipy/9781491922927/ch04.html">here</a>.

````python
f_s = 100  # Sampling rate, or number of measurements per second

t = X

fig, ax = plt.subplots()
ax.plot(t, y)
ax.set_xlabel('Time [s]')
ax.set_ylabel('Signal amplitude');
````

````python
from scipy import fftpack

Y = fftpack.fft(y)
freqs = fftpack.fftfreq(len(y)) * f_s

fig, ax = plt.subplots()

abs_magnitude = np.abs(Y)
relative_abs_magnitude = abs_magnitude/np.max(abs_magnitude)

ax.stem(freqs, relative_abs_magnitude)
ax.set_xlabel('Frequency in Hertz [Hz]')
ax.set_ylabel('Relative Magnitude')
ax.set_xlim(-f_s / 4, f_s / 4)
````

````python
np.argsort(Y)[::-1] # Use these indexes to find the frequencies with the highest magnitudes
````

There seems to be significant periodicity at ~10Hz. We now remove this signal from the signal. To do this, fit a curve of the form

$$f(t) = Asin(2pift) + constant$$

and then subtract f(t) from our original data.

````python
from scipy import optimize

def seasonal_func(x, a, b, c, d):
    return a * np.sin(2*3.14*freqs[31] * x + b) + c * np.sin(2*3.14*freqs[278] * x + d) 

params, params_covariance = optimize.curve_fit(seasonal_func, X, y,
                                               p0=[50, 70, 50, 50])

print(params)
````

````python
fig, ax = plt.subplots()
ax.plot(t, y)
ax.plot(t, seasonal_func(t, params[0], params[1], params[2], params[3]))
ax.set_xlabel('Time [s]')
ax.set_ylabel('Signal amplitude');
````

````python
y_deseasoned = y - seasonal_func(t, params[0], params[1], params[2], params[3])

plt.plot(y_deseasoned)
````

We should, in theory, have something resembling a stationary time series.

# Autocorrelation and Partial Autocorrelation

We now plot the autocorrelation function (ACF) and partial autocorrelation function (PACF) to help us figure out what type of ARIMA model to use. The difference between autocorrelation and partial autocorrelation is that partial autocorrelation removes the effects of intermediate time steps. This prevents situations where correlations extend back back into the past due to chained correlations. We do, however, need both.

Look <a href ="https://otexts.com/fpp2/non-seasonal-arima.html">here</a> for more information on using ACF and PACF plots to choose an ARIMA model.

````python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(y_deseasoned, lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(y_deseasoned, lags=40, ax=ax2)
````

# Auto-Fitting with Pmdarima
Alternatively, we can brute force various ARIMA models and choose the best based on the Akaike Information Criterion (AIC) or Bayesian Information Criterion (BIC). The package ```pmdarima``` does this for us.

````python
y_train = y_deseasoned[0:270]
y_test = y_deseasoned[270:]
````

````python
import pmdarima as pm

arima = pm.auto_arima(y_train, error_action='ignore', trace=1,
                      seasonal=False, m=1)

# #############################################################################
# Plot actual test vs. forecasts:
plt.plot(y_test, marker='x')
plt.plot(arima.predict(n_periods=np.shape(y_test)[0]))
plt.title('Actual test samples vs. forecasts')
plt.show()
````

````python
model_preds = arima.predict_in_sample()

plt.plot(y_train)
plt.plot(arima.predict_in_sample())
plt.title('Actual test samples vs. forecasts')
plt.show()
````

````python
arima.summary()
````

We now retrain the model in ```statsmodels``` using the sarimax class, as this has more functionality than the arima class (SARIMA can be reduced to ARIMA).

````python
training_mod = sm.tsa.SARIMAX(y_train, order=(4,1,1))
training_res = training_mod.fit()

# with updated data
mod = sm.tsa.SARIMAX(y_test, order=(4,1,1))
res = mod.filter(training_res.params)

insample = res.predict()

plt.plot(insample)
plt.plot(y_test)
````

# Building Final Model
After fitting an ARIMA model, we need to combine it with any previously identified trend and seasonality effects.

````python
def final_model(t, arima, seasonality, trend):
    
    final_preds = arima.predict(t, cv_samples = True), + seasonality(t) + trend(t)
    
    return final_preds

model_preds = final_model(t, arima,seasonal_func(t, params[0], params[1], params[2], params[3]), lambda x: 0)
````

````python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, max_error

# Other performance metrics
r_squared = r2_score(y_deseasoned[1:], model_preds)
rmse = np.sqrt(mean_squared_error(y_deseasoned[1:], model_preds))
mae = mean_absolute_error(y_deseasoned[1:], model_preds)
me = max_error(y_deseasoned[1:], model_preds)

print("r_squared: {}, rmse: {}, mae: {}, me: {}".format(r_squared, rmse, mae, me))
````
