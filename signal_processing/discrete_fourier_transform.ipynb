```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import pi
from scipy.fftpack import fft
```

```python
sample_rate = 1024
N = (2 - 0) * sample_rate
freq1 = 60
magnitude1 = 25
freq2 = 270
magnitude2 = 2
```

```python
time = np.linspace(0, 2, N)

waveform1 = magnitude1 * np.sin (2 * pi * freq1 * time)
waveform2 = magnitude2 * np.sin (2 * pi * freq2 * time)

noise = np.random.normal (0, 3, N)
time_data = waveform1 + waveform2 + noise

plt.plot (time [0:100], time_data [0:100])
plt.title ('Time Domain Signal')
plt.xlabel ('Time')
plt.ylabel ('Amplitude')
plt.show ()
```

```python
fft = np.fft.fft(time_data)[0:len(time_data)//2]
y = 2/N * np.abs(fft)
freq = np.linspace(0,int(sample_rate/2), int(N/2))

plt.plot(freq, y)
plt.title("Frequency Domain Signal")
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.show()
```
